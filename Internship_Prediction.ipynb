**Encoding**

Here, we can see that **Age** is already a numerical, no encoding is required.

The next column is **Gender**, on which we apply **Label** **Encoding** since it is a **binary** **categorical** **variable**.

```python
from sklearn.preprocessing import LabelEncoder
encoder1=LabelEncoder()
df['Gender']=encoder1.fit_transform(df['Gender'])
```

Next column is **Placement** **Status**, we will apply again **Label** **Encoding** Here.

```python
encoder2=LabelEncoder()
df['Placement_Status']=encoder2.fit_transform(df['Placement_Status'])
```

After completing **Placement Status** column, we move to the **Degree** column. **OneHotEncoding**  is applied since **Degree** is a **nominal** categorical feature with no inherent order.

```python
from sklearn.preprocessing import OneHotEncoder
encode3=OneHotEncoder()
```

```python
q=encode3.fit_transform(df[["Degree"]]).toarray()
q
```

```python
encoder3_df=pd.DataFrame(q,columns=encode3.get_feature_names_out())
df=pd.concat([df,encoder3_df],axis=1)
df.head()
```

Our next column is **Branch**, we will apply **One Hot Encoding** here.

```python
from sklearn.preprocessing import OneHotEncoder
encode4=OneHotEncoder(drop='first')
```

```python
r=encode4.fit_transform(df[["Branch"]]).toarray()
r
```

```python
encoder4_df=pd.DataFrame(r,columns=encode4.get_feature_names_out())
df=pd.concat([df,encoder4_df],axis=1)
df.head()
```

Now, we will drop categorical columns from the dataset

```python
df=df.drop(['Degree','Branch'],axis=1)
```

From here, we will separate **Independent** and **Dependent** variables

```python
X=df.drop("Placement_Status",axis=1)
y=df["Placement_Status"]
```

**Scaling** :

So, that all features are brought to the same scale.

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
X=scale.fit_transform(X)
```

Splitting Dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

**Choosing** **the** **best** **model**

**1. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
clas=LogisticRegression()
```

```python
clas.fit(X_train,y_train)
```

```python
y_pred=clas.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

Will calculate its metrics

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

**2. Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

**3. Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=150)
```

```python
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

Here, we can see that **Decision** **Tree** **Classifier** works better.
